**Performance at home vs away venue?**

```python
def lucky(matches,team_name):
    return matches[matches['winner']==team_name]['venue'].value_counts().nlargest(10)
```

```python
import plotly.graph_objs as go
```

**Who will Qualify as a first Finalist ? Season - 2024**

```python
matches['team1'].unique()
```

```python
# Team names
team1 = 'Royal Challengers Bangalore'
team2 = 'Chennai Super Kings'

# mathces between teams
matches_between_teams = matches[((matches['team1'] == team1) & (matches['team2'] == team2)) |
                           ((matches['team1'] == team2) & (matches['team2'] == team1))]

# total number of mathes betweeen two teams
total_matches = matches_between_teams.shape[0]

# Count the number of wins for each team
team1_wins = matches_between_teams[matches_between_teams['winner'] == team1].shape[0]
team2_wins = matches_between_teams[matches_between_teams['winner'] == team2].shape[0]

# Calculate the winning ratios
if total_matches > 0:
    team1_win_ratio = team1_wins / total_matches
    team2_win_ratio = team2_wins / total_matches
else:
    team1_win_ratio = 0
    team2_win_ratio = 0

print(f"Total matches between {team1} and {team2}: {total_matches}")
print(f"{team1} wins: {team1_wins}")
print(f"{team2} wins: {team2_wins}")
print(f"{team1} win ratio: {team1_win_ratio:.2f}")
print(f"{team2} win ratio: {team2_win_ratio:.2f}")
```
